```python
from sklearn.preprocessing import LabelEncoder
label_encoder_occupation = LabelEncoder()
label_encoder_car_use = LabelEncoder()
label_encoder_car_type = LabelEncoder()
label_encoder_revoked = LabelEncoder()
```

```python
label_encoder_occupation.fit(data['OCCUPATION'])
data['OCCUPATION'] = label_encoder_occupation.transform(data['OCCUPATION'])

label_encoder_car_use.fit(data['CAR_USE'])
data['CAR_USE'] = label_encoder_car_use.transform(data['CAR_USE'])

label_encoder_car_type.fit(data['CAR_TYPE'])
data['CAR_TYPE'] = label_encoder_car_type.transform(data['CAR_TYPE'])

label_encoder_revoked.fit(data['REVOKED'])
data['REVOKED'] = label_encoder_revoked.transform(data['REVOKED'])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
vars = ['OCCUPATION', 'TRAVTIME', 'CAR_USE', 'BLUEBOOK', 'CAR_TYPE', 'OLDCLAIM', 'CLM_FREQ', 'REVOKED', 'MVR_PTS', 'CLM_AMT', 'CAR_AGE']
X_train[vars] = scaler.fit_transform(X_train[vars])
```

```python
X_test[vars] = scaler.transform(X_test[vars])
X_test
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(min_samples_split=3)

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=101, min_samples_split=3)

from sklearn import svm
sv = svm.SVC(kernel='linear', gamma='scale', shrinking=False)

from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()

from sklearn.neighbors import KNeighborsClassifier
knc = KNeighborsClassifier(n_neighbors=5)

from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
Y_test_arr = Y_test.iloc[:,0:1].values

y_test_col = pd.DataFrame(Y_test_arr)
y_test_head = y_test_col.head(50)
```

```python
# dt.fit(X_train, Y_train)
# dt_pred = dt.predict(X_test)

rf.fit(X_train, Y_train)
rf_pred = rf.predict(X_test)

# sv.fit(X_train, Y_train)
# sv_pred = sv.predict(X_test)

# nb.fit(X_train, Y_train)
# nb_pred = nb.predict(X_test)

# knc.fit(X_train, Y_train)
# knc_pred = nb.predict(X_test)

lr.fit(X_train, Y_train)
lr_pred = lr.predict(X_test)
```
